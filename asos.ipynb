# **Importing required libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# **Importing dataset**

```python
# dataset: https://www.kaggle.com/datasets/trainingdatapro/asos-e-commerce-dataset-30845-products/data

df = pd.read_csv('products_asos.csv')
df.head(10)
```

# **Getting familiar with data**

```python
df.describe(include=['object'])
```

# **Data cleaning and wrangling**

```python
# clean data: remove null values
df.isnull().sum()
```

```python
df.dropna(how='all', inplace=True)
```

```python
df.isnull().sum()
```

```python
# replace image urls with image count
import ast

df['images_count'] = df['images'].apply(ast.literal_eval).str.len()
df.drop('images', axis=1, inplace=True)
df.head(10)

# drop url column
df.drop('url', axis=1, inplace=True)
```

```python
# add column to indicate if product is discounted
df['is_discounted'] = df['price'].str.contains('Now')
df.head()

# remove From and Now in price column
df['price'] = df['price'].replace(r'^(From  |Now  )', '', regex=True)
```

```python
# clean data: drop duplicated rows
df = df.drop_duplicates().reset_index(drop=True)

# clean data: uppercase color column
df['color'] = df['color'].str.upper()

df.head(10)
```

```python
# verify if name and category columns differ
print(pd.DataFrame(df['name'] != df['category']).value_counts())
df.loc[df['name'] != df['category']]
```

```python
# clean data: drop category column
df.drop('category', axis=1, inplace=True)
```

```python
# introduce dedicated columns for product brand and title
import re

# all uppercase words are treated as a brand name
# if 1st word is lowercase, then 1st word is treated as a brand name
# if all words are uppercase then brand is unknown
def extract_brand_title(name):
    if re.match(r'^[a-z]', name):
        # Split at the first space
        brand = name.split(' ', 1)[0]
        title = name.split(' ', 1)[1]
        return pd.Series([brand, title])
    else:
        # Match one or more consecutive title case or uppercase words at the start
        match = re.search(r'\b([a-z]\w*)\b', name)
        if match:
            i = match.start()
            brand = name[:i].strip()
            title = name[i:].strip()
            return pd.Series([brand, title])
        else:
            return pd.Series(['UNKNOWN', name])


df[['brand','title']] = df['name'].apply(extract_brand_title)
df.drop('name', axis=1, inplace=True)

# re-order columns
df = df[['sku', 'brand', 'title', 'price', 'is_discounted', 'color', 'size', 'description', 'images_count']]

# change column types
df['sku'] = df['sku'].astype('str')
df['price'] = df['price'].astype('float')
df['images_count'] = df['images_count'].astype('category')

df.head(10)
```

# **Data analysis**

```python
# find out top 10 brands based on the number of products
top_brands = df['brand'].value_counts().head(10)

sns.barplot(y=top_brands.index, x=top_brands.values, hue=top_brands.index, legend=False, palette='pastel')
plt.title('Top 10 Brands Based on Products Number', fontsize=16)
plt.xlabel('Number of Products', fontsize=14)
plt.ylabel('Brand', fontsize=14)
plt.show()
```

This visualization shows the top 10 brands ranked by number of products in a dataset. ASOS DESIGN dominates with nearly 4000 products, significantly outpacing the other brands. Other prominent brands include Topshop, adidas, and Miss Selfridge, each with around 1000–1500 products.



```python
# top 10 product colors based on the number of products
top_colors = df['color'].value_counts().head(10)

sns.barplot(y=top_colors.index, x=top_colors.values, hue=top_colors.index)
plt.title('Top 10 Product Colors', fontsize=16)
plt.xlabel('Number of Products', fontsize=14)
plt.ylabel('Color', fontsize=14)
plt.show()
```

The visualization displays the top 10 product colors by the number of products offered. Black products are the most popular, with over 6000 items. White, multi-color, and pink also rank highly, each with approximately 1500–2000 products.

The dataset dates back to 2023, and the visualization shows pink among the popular product colors. This aligns with the general trend of 2023, when Pantone’s Color of the Year was Viva Magenta — a bold and vibrant shade rooted in the red family with pinkish-purple undertones.




```python
# word cloud from product titles
from wordcloud import WordCloud

# Filter out non-string entries from the 'name' column
text = ' '.join([str(title) for title in df['title']])

# Generate the word cloud
wordcloud = WordCloud(background_color='white', width=600, height=400, max_words=150, max_font_size=50).generate(text)

# Plotting the word cloud
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Product Titles', fontsize=16)
plt.show()
```

This word cloud shows that dress styles like "mini dress," "midi dress," "maxi dress," and "long sleeve" appear most often in product names, meaning dresses are very common in this dataset. The most used colors are black, white, pink, blue, and green, which also confirms the fact these colors were popular in 2023.

Many words also describe details like "high neck," "puff sleeve," "crop," and "faux leather," which means lots of different dress shapes and features were available. The word "pink" is large, matching the trend for pink and magenta colors inspired by the Pantone Color of the Year, Viva Magenta.

```python
# ratio of discounted and non-discounted products

values = df['is_discounted'].value_counts()
labels = ['Non Discounted', 'Discounted']
explode = [0.1, 0]

plt.pie(values, labels=labels, labeldistance=0.4, autopct=lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*values.sum()), explode=explode)
plt.title('Ratio of Discounted & Non-discounted Products', fontsize=16)
plt.show()
```

This pie chart shows the proportion of discounted versus non-discounted products in the dataset. About 40% of products were sold at a discount, while approximately 60% were not discounted.

```python
# price binning

bins = np.linspace(df['price'].min(), df['price'].max(), 4)
group_names = ['Low', 'Medium', 'High']
pd.cut(df['price'], bins=bins, labels=group_names, include_lowest=True).value_counts()

```

```python
# price distribution
grouped_stats = df.groupby('is_discounted')['price'].agg(['mean', 'median'])

# Rename indexes for clarity
grouped_stats.index = ['Non Discounted', 'Discounted']

# swap rows and columns
grouped_stats.T.plot(kind='bar')

plt.title('Comparison of Price: Discounted vs Non Discounted')
plt.ylabel('Price')
plt.yticks(np.arange(0, 50, 5))
plt.legend(title='Product Type')
plt.show()
```

The bar chart shows that non-discounted products have both a higher mean and median price compared to discounted products. This means that, on average, items sold at full price are more expensive than those offered at a discount.

Median price for non-discounted products is around 37 GBP.

```python
values = df['images_count'].value_counts().sort_index()
labels = values.index.astype(str)

plt.figure(figsize=(7, 7))
wedges, texts = plt.pie(
    values,
    labels=None,
    explode=[0.4 if i == values.idxmax() else 0 for i in values.index]
)

legend_labels = [
    f"{label}: {count} ({count / values.sum() * 100:.2f}%)"
    for label, count in zip(labels, values)
]

plt.legend(wedges, legend_labels, title="Images per Product")
plt.title('Distribution of Products by Number of Images', fontsize=16)
plt.show()
```

The vast majority of products have 5 images, while other image counts occur much less frequently. This distribution confirms the common practice that having 4 to 6 images per product is considered optimal to effectively show the item.
